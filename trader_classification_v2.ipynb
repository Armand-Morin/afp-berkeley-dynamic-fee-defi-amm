```python
def get_profit(row):
    return (row['LVR'] - row['tcost_usd'])/row['amountUSD']

combined_df['profit'] = combined_df.apply(get_profit, axis=1)
combined_df.head()
```

```python
positive_count = combined_df[combined_df['profit'] > 0].shape[0]
total_count = combined_df.shape[0]

positive_count, total_count
```

```python
threshold = 0.1

trade_classification = {}

for sender in unique_senders:
    sender_df = combined_df[combined_df['sender'] == sender]
    positive_profit_count = sender_df[sender_df['profit'] > 0].shape[0]
    total_profit_count = sender_df.shape[0]
    
    if total_profit_count > 0:
        positive_profit_percentage = (positive_profit_count / total_profit_count) * 100
    else:
        positive_profit_percentage = 0
    
    classification = 'Arbitrageur' if positive_profit_percentage > threshold else 'Noise Trader'
    
    trade_classification[sender] = [f"{positive_profit_percentage:.2f}%", classification]
```

```python
list(trade_classification.items())[:5]
```

Data Preparation

```python
trade_class_df = pd.DataFrame(list(trade_classification.items()), columns=['Sender', 'Info'])
trade_class_df[['Positive Profit %', 'Classification']] = pd.DataFrame(trade_class_df['Info'].tolist(), index=trade_class_df.index)
trade_class_df.drop('Info', axis=1, inplace=True)
```

Visualization 1: Pie Chart of Classification Distribution

```python
classification_counts = trade_class_df['Classification'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(classification_counts, labels=classification_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Arbitrageurs vs. Noise Traders')
plt.show()
```

Visualization 2: Bar Chart of Positive Profit Percentage

```python
colors = trade_class_df['Classification'].map({'Arbitrageur': 'green', 'Noise Trader': 'red'})

plt.figure(figsize=(10, 7))
plt.bar(range(len(trade_class_df)), trade_class_df['Positive Profit %'].str.rstrip('%').astype('float'), color=colors)
plt.xlabel('Sender Index')
plt.ylabel('Positive Profit %')
plt.title('Positive Profit Percentage by Sender')
plt.xticks([])
plt.show()
```

Bubble Chart with Profit Percentage and Volume

```python
# Calculate the total traded volume
total_volume = combined_df['amountUSD'].sum()

# Calculate traded volume percentage for each sender
volume_percentage_per_sender = (combined_df.groupby('sender')['amountUSD'].sum() / total_volume) * 100

# Create a DataFrame from the trade_classification dictionary
trade_class_df = pd.DataFrame.from_dict(trade_classification, orient='index', columns=['Positive Trades %', 'Classification'])

# Convert 'Positive Trades %' from string percentage to numeric value
trade_class_df['Positive Trades %'] = trade_class_df['Positive Trades %'].str.rstrip('%').astype('float')

# Add the traded volume percentage to the trade_class_df DataFrame
trade_class_df = trade_class_df.merge(volume_percentage_per_sender.rename('Traded Volume %'), left_index=True, right_index=True)
```

```python
# Define colors for each classification
colors = {'Arbitrageur': 'green', 'Noise Trader': 'red'}
trade_class_df['Color'] = trade_class_df['Classification'].map(colors)

# Plotting
fig, ax = plt.subplots(figsize=(8, 5))

# Create bubble chart
scatter = ax.scatter(
    trade_class_df['Traded Volume %'], 
    trade_class_df['Positive Trades %'], 
    color=trade_class_df['Color'], 
    alpha=0.5, 
    edgecolors='w', 
    linewidth=0.5
)

# Apply labels and title
ax.set_xlabel('Percentage of Total Traded Volume')
ax.set_ylabel('Percentage of Positive Trades')
ax.set_title('Trader Analysis: Volume and Profitability')

# Legend for classifications
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Arbitrageur', markerfacecolor='green', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Noise Trader', markerfacecolor='red', markersize=10)
]
ax.legend(handles=legend_elements, title="Classification")

plt.show()
```
